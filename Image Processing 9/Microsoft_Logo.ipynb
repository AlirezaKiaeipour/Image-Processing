```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import ImageFont, ImageDraw, Image
```

```python
img = np.ones((630,1420,3),np.uint8)*74
img[210:310,210:310] = (246,83,20)
img[320:420,210:310] = (0,161,241)
img[210:310,320:420] = (124,187,0)
img[320:420,320:420] = (255,187,0)
font = ImageFont.truetype("font/seguisb.ttf",160)
image = Image.fromarray(img)
draw = ImageDraw.Draw(image)
draw.text((480,195),"Microsoft",font=font,fill=(255,255,255))
img = np.array(image)
plt.imshow(img)
plt.imsave("logo.jpg",img)
```
