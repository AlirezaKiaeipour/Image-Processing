```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("img/rubix.png")
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
row ,col = img_rgb.shape[:2]
for i in range (row):
    for j in range (col):
        if img_rgb[i,j][0]>50 and img_rgb[i,j][1]>50 and img_rgb[i,j][2]<50:
            img_rgb[i,j]=(0,0,255)
        elif img_rgb[i,j][0]<50 and img_rgb[i,j][1]>50 and img_rgb[i,j][2]>50:
            img_rgb[i,j] = (255,0,0)
        elif img_rgb[i,j][0]>50 and img_rgb[i,j][1]<50 and img_rgb[i,j][2]>50:
            img_rgb[i,j]=(0,255,0)
```
