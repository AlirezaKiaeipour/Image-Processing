```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("img/Carrot.jpg")
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
row ,col = img_rgb.shape[:2]
for i in range(400,row):
    for j in range(col):
        if img_rgb[i,j][0]>20 and img_rgb[i,j][1]>20 and img_rgb[i,j][2]<100:
            img_rgb[i,j]=(0,255,0)
plt.imshow(img_rgb)
```

```python
img = cv2.imread("img/Carrot.jpg")
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
R,G,B = cv2.split(img_rgb)
img_res = cv2.merge((G,R,B))
plt.imshow(img_res)
```
