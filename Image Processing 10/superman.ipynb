```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("img/SuperMan.jpg")
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
row ,col = img.shape[:2]
img_background = cv2.imread("img/backgrund1.jpg")
img_background_rgb = cv2.cvtColor(img_background,cv2.COLOR_BGR2RGB)
img_background_rgb = cv2.resize(img_background_rgb,(col,row))
plt.imshow(img_background_rgb)
```

```python
img_hsv = cv2.cvtColor(img_rgb,cv2.COLOR_RGB2HSV)
H,S,V = cv2.split(img_hsv)
```

```python
for i in range(row):
    for j in range(col):
        if 35 < H[i,j] < 75 and V[i,j]>85:
            img_rgb[i,j] = img_background_rgb[i,j]
```
